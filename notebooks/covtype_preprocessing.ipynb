```python
X_full = covtype_full.drop(columns=[54])
y_full = covtype_full[54]
y_full = y_full - 1  # target start from 0
X_full.head(5)
```

```python
from sklearn.model_selection import train_test_split
_, X, _, y = train_test_split(
    X_full, y_full,
    test_size=50000,
    stratify=y_full,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
X_train.shape
```

```python
X_train.to_csv(data_dir + 'covtype_X_train_processed.csv', index=False)
X_test.to_csv(data_dir + 'covtype_X_test_processed.csv', index=False)
y_train.to_csv(data_dir + 'covtype_y_train_processed.csv', index=False)
y_test.to_csv(data_dir + 'covtype_y_test_processed.csv', index=False)
```
