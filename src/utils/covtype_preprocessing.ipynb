```python
X = covtype.drop(columns=[54])
y = covtype[54]
y = y-1 # target start from 0
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y ,random_state=42)
```

```python
X_train.to_csv('covtype_X_train.csv',index=False)
X_test.to_csv('covtype_X_test.csv',index=False)
y_train.to_csv('covtype_y_train.csv',index=False)
y_test.to_csv('covtype_y_test.csv',index=False)
```
